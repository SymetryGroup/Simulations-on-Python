# Problema 1 / Gráfico del campo eléctrico

```python
import numpy as np

import matplotlib.pyplot as plt 


x = np.linspace(0,7,20)
y = np.linspace(0,7,20)


(x,y) = np.meshgrid(x,y)


x1 = np.linspace(0.1,7,200)
y1 = np.linspace(0.1,7,200)


(x1,y1)=np.meshgrid(x1,y1)

a = 4
b = 2.5 

pot = 1/((x1-a)**2+(y1-b)**2)**0.5-1/((x1-a)**2+(y1+b)**2)**0.5
-1/((x1+a)**2+(y1-b)**2)**0.5+1/((x1+a)**2+(y1+b)**2)**0.5


Ex = (x-a)/((x-a)**2+(y-b)**2)**1.5-(x-a)/((x-a)**2+(y+b)**2)**1.5-(x+a)/((x+a)**2+(y-b)**2)**1.5+(x+a)/((x+a)**2+(y+b)**2)**1.5

Ey = (y-b)/((x-a)**2+(y-b)**2)**1.5-(y+b)/((x-a)**2+(y+b)**2)**1.5-(y-b)/((x+a)**2+(y-b)**2)**1.5+(y+b)/((x+a)**2+(y+b)**2)**1.5

color = np.sqrt(Ex**2+Ey**2)

Ex = Ex/color

Ey = Ey/color

plt.quiver(x,y,Ex,Ey,np.log(color),pivot = 'middle',cmap='plasma')
plt.colorbar()
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Campo electrico 'r'$\vec{E}(x,y)$')
plt.show()
```

# Superficies/Lineas equipotenciales

```python
plt.quiver(x,y,Ex,Ey,np.log(color),pivot = 'middle',cmap='plasma')
plt.colorbar()
c = plt.contour(x1,y1,pot, levels=[0.15,0.25,0.5,1,1.5], colors='blue')
plt.clabel(c,inline=True)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Campo electrico 'r'$\vec{E}(x,y)$ '+'con sus lineas equipotenciales correspondientes' )
plt.show()
```

# Problema 2 / Gráfico del campo eléctrico

```python
x = np.linspace(-15,15,30)
y = np.linspace(-15,15,30)


(x,y) = np.meshgrid(x,y)


Ex = (x-10)/((x-10)**2+(y**2))**1.5+(x+10)/((x+10)**2+y**2)**1.5-0.125*(x-0.15)/((x-0.15)**2+y**2)**1.5-0.125*(x+0.15)/((x+0.15)**2+y**2)**1.5

Ey= y/((x-10)**2+(y**2))**1.5+y/((x+10)**2+y**2)**1.5-0.125*y/((x-0.15)**2+y**2)**1.5-0.125*y/((x+0.15)**2+y**2)**1.5

color = np.sqrt(Ex**2+Ey**2)

Ex = Ex/color

Ey = Ey/color

f = np.linspace(-1.25,1.25,200)

d1 = np.sqrt(1.25**2-f**2)
d2 = -np.sqrt(1.25**2-f**2)


plt.quiver(x,y,Ex,Ey,np.log(color),pivot = 'middle',cmap='plasma')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.plot(f,d1,color='black')
plt.plot(f,d2,color='black')
plt.title('Campo electrico 'r'$\vec{E}(x,y)$')
plt.colorbar()
```
